```python
import xmltodict

with open('C:/Users/luke/Downloads/Dictionary.xml', encoding="utf-8") as fd:
    doc = xmltodict.parse(fd.read())
```

```python
doc['dictionary']['chapter'][0]['@chapterid']
```

```python
dictchp = dict()
```

```python
chprng = len(doc['dictionary']['chapter'])
```

```python
for chp in range(chprng):
    try:
        chpid = (doc['dictionary']['chapter'][chp]['@chapterid'])
        chptitle =(doc['dictionary']['chapter'][chp]['chaptertitle'])
        #print(doc['dictionary']['chapter'][chp]['section'])
        dictchp.update({chpid : dict({'chaptertitle' : chptitle})})
    except KeyError:
        pass
    #print(doc['dictionary']['chapter'][chp])
```
